```python
import rake
import operator
import io 
import six
path="C:/Users/Dell/Downloads/RAKE-tutorial-master/data/stoplists/SmartStoplist.txt"
rake_obj=rake.Rake(path,3,3,3) #Each word has at least 3 characters
 #Each phrase has at most 3 words
 #Each keyword appears in the text at least 3 times
```

```python
# importing required modules
import PyPDF2

# creating a pdf file object
pdfFileObj=open('C:/Users/Dell/Downloads/JavaBasics-notes.pdf', 'rb')

# creating a pdf reader object
pdfReader=PyPDF2.PdfFileReader(pdfFileObj)

# printing number of pages in pdf file
print(pdfReader.numPages)

# creating a page object
num_pages = pdfReader.numPages
count = 0
text = ""
#The while loop will read each page
while count < num_pages:
    pageObj=pdfReader.getPage(count)
    count=count+1
    text=text+pageObj.extractText()
    print(text)
# closing the pdf file object
pdfFileObj.close()
```

```python
keywords=rake_obj.run(text)
print("Keywords:", keywords)
```

```python
filenames = ['C:/Users/Dell/Downloads/JavaBasics-notes.pdf']   
results={}
for filename in filenames:
    with open(filename, 'r') as fp:
        results[filename] = rake_obj.run(text)
results
```

```python
import pandas as pd
words_with_weights=pd.DataFrame({"keyword-weights":keywords})
words_with_weights.to_csv('predictions.csv', index=False, header=True)
words_with_weights
```
